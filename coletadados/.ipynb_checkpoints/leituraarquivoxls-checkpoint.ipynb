```python
# ler arquivo xls
df_tsa = pd.read_excel('claims-2014.xls', sheet_name = 'interactive_report')
```

```python
# visualizar estrutura dataframe
df_tsa.info()
```

```python
# converter '-' em zero na coluna Close Amount
df_tsa.loc[df_tsa['Close Amount'] == '-', 'Close Amount'] = 0
```

```python
# transformar o campo Close Amount em valor
df_tsa['Close Amount'] = df_tsa['Close Amount'].astype('float')
```

```python
# visualizar distribuição de valores de perda
df_tsa['Close Amount'].hist()
```

```python
# visualizar métricas dos valores de perda
df_tsa['Close Amount'].describe()
```
