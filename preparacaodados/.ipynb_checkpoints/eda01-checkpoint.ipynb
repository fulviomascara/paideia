```python
# importar bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# setar configurações
%matplotlib inline
pd.options.display.max_rows = 6000
pd.options.display.max_columns = 999
```

```python
# carregar o dataframe df_brcities com o arquivo BRAZIL_CITIES.csv
df_brcities = pd.read_csv("BRAZIL_CITIES.csv", sep = ";")
```

```python
# Gráfico de Barras por Característica de Cidade
df_brcities.RURAL_URBAN.value_counts().plot(kind='bar', colors='green')
```

```python
# Gráfico de Barras Quantidade de Cidades por Estado
plt.figure(figsize=(10,8))
ax = df_brcities.plot.bar(x='STATE', y='COMP_TOT', rot=270) 
```

```python
# Histograma por GDP_CAPITA
sns.distplot(df_brcities.GDP_CAPITA, color="seagreen")
```

```python
# Box Plot
plt.figure(figsize=(10,8))
sns.boxplot(x="GDP_CAPITA", y="GVA_MAIN", data=df_brcities)
```
